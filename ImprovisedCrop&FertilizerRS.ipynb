Importing necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the Dataset

```python
crop=pd.read_csv('Crop_recommendation.csv')
```

Heatmap

```python
# Function to generate the heatmap
def generate_heatmap(data):
    """
    Generates a heatmap for crop suitability based on key features.
    """
    # Grouping data by 'label' and calculating the mean for each crop type
    grouped_data = data.groupby('label').mean()

    # Selecting relevant columns for the heatmap
    heatmap_data = grouped_data[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]

    # Creating the heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(
        heatmap_data,
        annot=True,
        fmt=".2f",
        cmap="YlGnBu",
        linewidths=0.5,
        cbar_kws={'label': 'Average Value'}
    )
    plt.title("Crop Suitability Heatmap", fontsize=16)
    plt.ylabel("Crop Type", fontsize=12)
    plt.xlabel("Soil and Weather Conditions", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Call the function after loading your dataset
generate_heatmap(crop)
```
